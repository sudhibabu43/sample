```python
number=int(input("Enter a number :"))
sum=0
while number!=0:
  rem=number%10
  sum=sum+rem
  number=number//10
print(sum)
```

```python
s=input("Enter a String\n")
letters=s.split()
vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
for letter in letters:
  vowelcount=0
  for i in range(0,len(letter)):
    if letter[i] in vowels:
      vowelcount=vowelcount+1
  print(vowelcount)
```
