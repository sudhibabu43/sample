```python
print("Palindrom\n")
s=input("Enter a sentence :\n").lower()
new=""
for i in s:
  if(i!=" "):
    new=new+i
rev=new[::-1]
flag=0
for i in range(len(rev)):
  if(new[i]==rev[i]):
    flag=0
  else:
    flag=1
    break
if (flag==1):
  print("Not a palindrom")
else:
  print("Is a palindrom")
```

```python
print("Palindrom\n")
s=input("Enter a sentence :\n").lower().replace(" ","")
if(s==s[::-1]):
    print("Is a palindrom")
else:
  print("Not a palindrom")

```
