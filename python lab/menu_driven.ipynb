```python
while(1):
  num1=int(input("Enter the first number :"))
  num2=int(input("Enter the second number "))
  print("\tMenu\n")
  print("1. And operation\n2. Or operation\n3. Not\n4. Xor\n5. exit")
  n=int(input("Choose the option :"))
  match n:
      case 1:print("Answer is :",num1 and num2)
      case 2:print("Answer is :",num1 or num2)
      case 3:print("Answer is :",num1|num2)
      case 4:print("Answer is :",num1^num2)
      case 5:exit(0)
      case _:print("Invalid choice ")
```
